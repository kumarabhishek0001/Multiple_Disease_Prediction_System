```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

<h3>Data Collection and preprocessing

```python
parkinsons_data = pd.read_csv("Data\parkinsons.csv")
parkinsons_data
```

```python
parkinsons_data.sample(5)
```

```python
parkinsons_data.describe().T
```

```python
#distribution of target
parkinsons_data['status'].value_counts()
```

```python
#splitting model
X = parkinsons_data.drop(columns=['name', 'status'], axis=1)
y = parkinsons_data['status']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify= y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
#standaising data
scaler = StandardScaler()
X_train =scaler.fit_transform(X_train)
```

```python
X_test =scaler.fit_transform(X_test)
```

```python
# TRAINING MODEL
model = svm.SVC(kernel='linear')
```

```python
model.fit(X_train, y_train)
```

```python
#Model evaluation
X_train_prediction = model.predict(X_train)
```

```python
train_data_accuracy = accuracy_score(y_train, X_train_prediction)
print('training data accuracy score is : ', train_data_accuracy  )
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(y_test, X_test_prediction)
print('testing data accuracy score is : ', test_data_accuracy  )
```
